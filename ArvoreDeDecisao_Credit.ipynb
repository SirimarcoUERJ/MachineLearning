# **Árvore de Decisão**

```python
from sklearn.tree import DecisionTreeClassifier
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount("/content/drive")
```

```python
arq = "/content/drive/MyDrive/Colab Notebooks/Machine Learn/data/credit.pkl"
with open(arq, "rb") as f:
  x_treino, y_treino, x_teste, y_teste = pkl.load(f)
```

```python
x_treino.shape, y_treino.shape
```

```python
x_teste.shape, y_teste.shape
```

```python
arvore_credit = DecisionTreeClassifier(criterion="entropy", random_state=0)
arvore_credit.fit(x_treino, y_treino)
```

```python
arvore_credit.feature_importances_ #quanto maior o valor da variavel maior o ganho de informação. [income, age, loan]
```

```python
previsoes = arvore_credit.predict(x_teste)
previsoes
```

```python
y_teste
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

accuracy_score(y_teste, previsoes)
```

```python
confusion_matrix(y_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(arvore_credit)
cm.fit(x_treino, y_treino)
cm.score(x_teste, y_teste)
```

```python
print(classification_report(y_teste, previsoes))
```

```python
from sklearn import tree
previsores = ["income", "age", "loan"]
classes = ["paga", "não paga"]

plt.figure(figsize=(20,20))
tree.plot_tree(arvore_credit, feature_names=previsores,  class_names=classes,filled=True);
plt.show()
```
